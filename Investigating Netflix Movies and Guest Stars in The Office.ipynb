```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [11, 7]

train_data = pd.read_csv('datasets/office_episodes.csv')
train_data.head()

```

```python
colors=[]
for lab,row in train_data.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append('red')
    elif row['scaled_ratings'] >=0.25 and row['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif row['scaled_ratings'] >=0.50 and row['scaled_ratings'] <0.75:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')

colors[0:20]    
```

```python
size=[]
for ind,row in train_data.iterrows():
    if row['has_guests'] == True:
        size.append(250)
    else:
        size.append(25)

size[0:5]
```

```python
train_data['colors'] = colors
train_data['size'] = size
```

```python
non_guest = train_data[train_data['has_guests']==False]
guest = train_data[train_data['has_guests']==True]
```

```python
fig = plt.figure()
plt.scatter(non_guest['episode_number'],non_guest['viewership_mil'],
            c=non_guest['colors'],s=non_guest['size'])
plt.scatter(guest['episode_number'],guest['viewership_mil'],
            c=guest['colors'],s=guest['size'],marker='*')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```
